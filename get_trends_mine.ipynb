```python
import pyautogui
import time
pyautogui.FAILSAFE = False
import pandas as pd

time.sleep(5)

df = pd.read_csv('urls.csv')
urls = list(df['Values'])
folder_path = 'C:\\Users\\chait\\Desktop\\gtrends'

# Loop through each URL
n = 1
for url in urls:
    pyautogui.hotkey('ctrl', 'l')  # Go to the URL bar
    pyautogui.write(url)
    pyautogui.press('enter')  # Hit Enter to visit the website
    
    time.sleep(3)
    
    pyautogui.click(1521, 626)
    
    pyautogui.hotkey('ctrl', 'l')  # Go to the URL bar
    pyautogui.write(folder_path)    
    pyautogui.press('enter')
    time.sleep(5)
    pyautogui.hotkey('alt', 'n')
    f_name = 'trend_'+str(n)
    pyautogui.write(f_name)
    n+=1
    
    pyautogui.press('enter')
    
    time.sleep(3)
```
